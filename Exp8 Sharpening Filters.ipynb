# Gaussian Filter

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt

img = cv2.imread('spideyS1.jpg',cv2.COLOR_BGR2RGB)

g_img = cv2.GaussianBlur(img, (3,3), 2)
g_img1 = cv2.GaussianBlur(img, (5,5), 2)
g_img2 = cv2.GaussianBlur(img, (7,7), 2)

cv2.imshow("Original", img)
cv2.imshow('Gaussian Filtered',g_img)
cv2.imshow('Gaussian Filtered 1',g_img1)
cv2.imshow('Gaussian Filtered 2',g_img2)

cv2.waitKey(0) 
cv2.destroyAllWindows()
```

# Laplacian Filter

```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt
ddepth = cv.CV_16S

img = cv.imread('spideyS1.jpg')
g_img = cv.GaussianBlur(img, (3,3), 0)
g_img = cv.cvtColor(g_img, cv.COLOR_BGR2GRAY)


l_img = cv.Laplacian(g_img, ddepth, ksize=3)
l_img1 = cv.Laplacian(g_img, ddepth, ksize=5)
l_img2 = cv.Laplacian(g_img, ddepth, ksize=7)

cv.imshow("Original", img)
cv.imshow('Laplacian Filtered',l_img)
cv.imshow('Laplacian Filtered 1',l_img1)
cv.imshow('Laplacian Filtered 2',l_img2)

cv.waitKey(0) 
cv.destroyAllWindows()
```
