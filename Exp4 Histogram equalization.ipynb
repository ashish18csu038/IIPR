# Log Transformation

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
   

image = cv2.imread('spidey.jpg') 
```

```python
cv2.imshow("img",image)
cv2.waitKey(0)
```

```python

   

c = 255 / np.log(1 + np.max(image)) 
s = c * (np.log(image + 1)) 
   

s = np.array(s, dtype = np.uint8) 
   
 
plt.imshow(image) 
plt.show() 
plt.imshow(s) 
plt.show() 
```

# Tranformation

```python
    
gamma1 = np.array(255*(image / 255) ** 0.01, dtype = 'uint8') 
gamma2 = np.array(255*(image / 255) ** 0.02, dtype = 'uint8') 
gamma3 = np.array(255*(image / 255) ** 0.5, dtype = 'uint8') 
gamma4 = np.array(255*(image / 255) ** 1.0, dtype = 'uint8') 
gamma5 = np.array(255*(image / 255) ** 2.0, dtype = 'uint8') 
gamma6 = np.array(255*(image / 255) ** 3.0, dtype = 'uint8') 
gamma7 = np.array(255*(image / 255) ** 4.0, dtype = 'uint8') 
plt.imshow(gamma1) 
plt.show() 
plt.imshow(gamma2) 
plt.show() 
plt.imshow(gamma3) 
plt.show() 
plt.imshow(gamma4) 
plt.show() 
plt.imshow(gamma5) 
plt.show() 
plt.imshow(gamma6) 
plt.show()
plt.imshow(gamma7) 
plt.show()
```

# Histogram Equalization

```python
img = cv2.imread('spidey.jpg')
```

```python
plt.imshow(img) 
plt.show() 
```

```python
#Grayscale or blue channel
 
hist = cv2.calcHist([img],[0],None,[256],[0,256]) 
  

plt.plot(hist) 
plt.show()
```

```python
#green channel

hist = cv2.calcHist([img],[1],None,[256],[0,256]) 
plt.plot(hist)
```

```python
#red channel

hist = cv2.calcHist([img],[2],None,[256],[0,256]) 
plt.plot(hist)
```

```python
# calculating histogram using numpy


plt.hist(img.ravel(),256,[0,256]); plt.show()
```

```python
#rgb
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python

mask = np.zeros(img.shape[:2], np.uint8)
mask[0:500, 250:750] = 255
masked_img = cv2.bitwise_and(img,img,mask = mask)

hist_full = cv2.calcHist([img],[0],None,[256],[0,256])
hist_mask = cv2.calcHist([img],[0],mask,[256],[0,256])

plt.subplot(221), plt.imshow(img, 'gray')
plt.subplot(222), plt.imshow(mask,'gray')
plt.subplot(223), plt.imshow(masked_img, 'gray')
plt.subplot(224), plt.plot(hist_full), plt.plot(hist_mask)
plt.xlim([0,256])

plt.show()
```

# Contrast Stretching

```python
import cv2
import numpy as np

# Read the image
img1 = cv2.imread('spidey.jpg',0)
print

# Create zeros array to store the stretched image
minmax_img = np.zeros((img1.shape[0],img1.shape[1]),dtype = 'uint8')

# Loop over the image and apply Min-Max formulae
for i in range(img1.shape[0]):
    for j in range(img1.shape[1]):
        minmax_img[i,j] = 255*(img1[i,j]-np.min(img1))/(np.max(img1)-np.min(img1))

# Displat the stretched image
cv2.imshow('Minmax',minmax_img)
cv2.waitKey(0)
```
