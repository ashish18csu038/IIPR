```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

org = cv.imread('spideyS1.jpg', 0)
img = cv.dft(np.float32(org), flags = cv.DFT_COMPLEX_OUTPUT)
img_shift = np.fft.fftshift(img)

magnitude_spectrum = 20*np.log(cv.magnitude(img_shift[:,:,0],img_shift[:,:,1]))

cv.imshow("original", org)
cv.imshow("Magnitude spectrum", magnitude_spectrum)
cv.waitKey(0) 
cv.destroyAllWindows()
```

```python
rows, cols = org.shape
crow,ccol = int(rows/2), int(cols/2)
mask=np.zeros((rows,cols,2),np.uint8)
mask[crow-30:crow+30,ccol-30:ccol+30]=1
```

```python
fshift=img_shift*mask
f_ishift = np.fft.ifftshift(fshift)
img_back = cv.idft(f_ishift)
img_back = cv.magnitude(img_back[:,:,0],img_back[:,:,1])

cv.imshow("original", org)
cv.imshow("Result", img_back)
cv.waitKey(0) 
cv.destroyAllWindows()
```
