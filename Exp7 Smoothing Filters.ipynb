# Average Filter

```python
import cv2 
import numpy as np 

img = cv2.imread('spideyS.jpg', 0) 
  
m, n = img.shape 
 
mask = np.ones([3, 3], dtype = int) 
mask = mask / 9
   
img_new = np.zeros([m, n]) 
  
for i in range(1, m-1): 
    for j in range(1, n-1): 
        temp = img[i-1, j-1]*mask[0, 0]+img[i-1, j]*mask[0, 1]+img[i-1, j + 1]*mask[0, 2]+img[i, j-1]*mask[1, 0]+ img[i, j]*mask[1, 1]+img[i, j + 1]*mask[1, 2]+img[i + 1, j-1]*mask[2, 0]+img[i + 1, j]*mask[2, 1]+img[i + 1, j + 1]*mask[2, 2] 
         
        img_new[i, j]= temp 
          
img_new = img_new.astype(np.uint8) 
cv2.imwrite('blurred.tif', img_new)
cv2.imshow("Original", img)
cv2.imshow('Average',img_new)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

# Median Filter

```python
import cv2 
import numpy as np 

img_noisy1 = cv2.imread('spideyS.jpg', 0) 
  
m, n = img_noisy1.shape 
   
img_new1 = np.zeros([m, n]) 
  
for i in range(1, m-1): 
    for j in range(1, n-1): 
        temp = [img_noisy1[i-1, j-1], 
               img_noisy1[i-1, j], 
               img_noisy1[i-1, j + 1], 
               img_noisy1[i, j-1], 
               img_noisy1[i, j], 
               img_noisy1[i, j + 1], 
               img_noisy1[i + 1, j-1], 
               img_noisy1[i + 1, j], 
               img_noisy1[i + 1, j + 1]] 
          
        temp = sorted(temp) 
        img_new1[i, j]= temp[4] 
  
img_new1 = img_new1.astype(np.uint8)
cv2.imshow("Original", img)
cv2.imshow('Median Filtered',img_new1)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

# Blurred

```python
img = cv2.imread('spideyS.jpg',cv2.COLOR_BGR2RGB)

blur = cv2.blur(img, (25,25))

cv2.imshow("Original", img)
cv2.imshow('Blurred',blur)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

# Median Blurred

```python
Mblur = cv2.medianBlur(img, 25)

cv2.imshow("Original", img)
cv2.imshow('Median Blurred',Mblur)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

# Box Filter Blurred

```python
Fblur = cv2.boxFilter(img,-1,(25,25), normalize = True)

cv2.imshow("Original", img)
cv2.imshow('Box Filter Blurred',Fblur)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

# Min Filter

```python
img = cv2.imread('spideyS1.jpg',cv2.COLOR_BGR2RGB)

kernel = np.ones((5,5), np.uint8)
min = cv2.erode(img,kernel,iterations=1)

kernel1 = np.ones((9,9), np.uint8)
min1 = cv2.erode(img,kernel1,iterations=1)

kernel2 = np.ones((11,11), np.uint8)
min2 = cv2.erode(img,kernel2,iterations=1)

cv2.imshow("Original", img)
cv2.imshow('Min Filter',min)
cv2.imshow('Min Filter1',min1)
cv2.imshow('Min Filter2',min2)

cv2.waitKey(0) 
cv2.destroyAllWindows()
```

# Max Filter

```python
img = cv2.imread('spideyS1.jpg',cv2.COLOR_BGR2RGB)

kernel = np.ones((5,5), np.uint8)
max = cv2.dilate(img,kernel,iterations=1)

kernel1 = np.ones((9,9), np.uint8)
max1 = cv2.dilate(img,kernel1,iterations=1)

kernel2 = np.ones((11,11), np.uint8)
max2 = cv2.dilate(img,kernel2,iterations=1)

cv2.imshow("Original", img)
cv2.imshow('Max Filter',max)
cv2.imshow('Max Filter1',max1)
cv2.imshow('Max Filter2',max2)

cv2.waitKey(0) 
cv2.destroyAllWindows()
```
