# Convolution Filter

```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage

image = cv2.imread('spideyS.jpg', 0)
kernel=np.array([[1,1,1],[1,1,0],[1,0,0]])
dist_conv = ndimage.convolve(image, kernel, mode='constant', cval=1.0)
cv2.imshow('conv', dist_conv)
cv2.waitKey()
cv2.destroyAllWindows()
```

# Correlation Filter

```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage

image = cv2.imread('spideyS.jpg', 0)
kernel=np.array([[1,1,1],[1,1,0],[1,0,0]])
dist_corr = cv2.filter2D(image, -1, kernel)
cv2.imshow('corr', dist_corr)
cv2.waitKey()
cv2.destroyAllWindows()
```
