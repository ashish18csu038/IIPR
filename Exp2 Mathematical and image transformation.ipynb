```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image1 = np.zeros((400, 400), dtype="uint8")
cv2.rectangle(image1, (100, 100), (250, 250), 255, -1)
cv2.imshow('First Image', image1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image2=np.zeros((400, 400), dtype="uint8")
cv2.circle(image2, (150, 150), 90, 255, -1)
cv2.imshow('Second image', image2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#and
image3=cv2.bitwise_and(image1, image2)
cv2.imshow('AND Image', image3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#0r
image4=cv2.bitwise_or(image1, image2)
cv2.imshow('OR Image', image4)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#BitNot
image5=cv2.bitwise_not(image1, image2)
cv2.imshow('NOT Image', image5)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Xor
image6=cv2.bitwise_xor(image1, image2)
cv2.imshow('XOR Image', image6)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
